# **Лабораторная работа №1: Введение в машинное зрение**

# Выгрузка необходимых библиотек

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Загрузите изображение с использованием библиотеки OpenCV.

```python
image = cv2.imread('images.jpeg')
```

 # 2. Отобразите изображение с помощью Matplotlib.

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

# 3. Преобразуйте изображение в градации серого и отобразите его.

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # Преобразовал в градации серого
plt.imshow(gray_image, cmap='gray')
plt.title('Gray version')
plt.axis('off')
plt.show()
```

# 4. Преобразуйте изображение в задании 1 из RGB в HSV.


```python
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt.imshow(cv2.cvtColor(hsv_image, cv2.COLOR_BGR2RGB))
plt.title('HSV')
plt.axis('off')
plt.show()

# Инверсия изображения сваязана с переходом из hsv в rgb по стандарту Matplotlib
```

# 5. Отобразите каждый канал (Hue, Saturation, Value) отдельно.

```python
hue, saturation, value = cv2.split(hsv_image)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(hue, cmap='hsv')
plt.title('Канал Hue - оттенок')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(saturation, cmap='gray')
plt.title('Канал Saturation - насыщенность')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(value, cmap='gray')
plt.title('Канал Value - яркость')
plt.axis('off')

plt.show()
```

# 6. Примените поворот изображения (в задании 1) на 45 градусов.

```python
(h, w) = image.shape[:2]
center = (w // 2, h // 2)
matrix = cv2.getRotationMatrix2D(center, 45, 1.0) #  Матрица поворота на 45 градусов
rotated_image = cv2.warpAffine(image, matrix, (w, h)) # Поворот изображения

plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title('Изображение на 45 градусов')
plt.axis('off')
plt.show()
```

# 7. Масштабируйте изображение (в задании 1) в 2 раза.

```python
scaled_image = cv2.resize(image, None, fx=6, fy=2) # Растянул по оси x в 6 раз для наглядности

plt.imshow(cv2.cvtColor(scaled_image, cv2.COLOR_BGR2RGB))
plt.title('Масштабируемое изображение')
plt.axis('off')
plt.show()
```

# 8. Примените сдвиг изображения (в задании 1) по горизонтали на 50 пикселей.

```python
M = np.float32([[1, 0, 50], [0, 1, 0]])  # Сдвиг по X на 50 пикселей
shifted_image = cv2.warpAffine(image, M, (w, h)) # сдвиг к изображению

plt.imshow(cv2.cvtColor(shifted_image, cv2.COLOR_BGR2RGB))
plt.title('Сдвинутое изображение')
plt.axis('off')
plt.show()
```

1. Основные задачи машинного зрения:

   • Обнаружение объектов: Выявление и локализация объектов на изображении.

   • Сегментация: Разделение изображения на значимые области или объекты.

   • Распознавание образов: Идентификация и классификация объектов или сцен.

   • Трacking (отслеживание): Мониторинг перемещения объектов в последовательности изображений.

   • Анализ видео: Извлечение информации из видеопотока.

   • Реконструкция 3D: Создание трехмерных моделей из двумерных изображений.

   • Оптическое распознавание символов (OCR): Преобразование текста на изображениях в редактируемый текст.


---



2. Разница между цветовыми моделями RGB и HSV:

   • RGB (Red, Green, Blue): Это аддитивная цветовая модель, где цвета создаются путем смешивания красного, зеленого и синего света. Цвета описываются как комбинации этих трех компонентов в диапазоне от 0 до 255.

   • HSV (Hue, Saturation, Value): Это цветовая модель, которая описывает цвет в терминах оттенка (Hue), насыщенности (Saturation) и яркости (Value). Она более интуитивна для восприятия человеком, так как позволяет легче изменять цветовые характеристики, например, изменять оттенок или насыщенность без изменения яркости.

Примечание:  BGR - также аддитивная цветовая модель, аналогичная RGB, но порядок компонентов изменен, используется в библиотеках компьютерного зрения, таких как OpenCV. Это связано с тем, что многие форматы изображений хранят данные в порядке BGR.


---



3. Проблемы при обработке изображений:

   • Шум: Нежелательные артефакты, которые могут ухудшать качество изображения.

   • Изменения освещения: Изменения условий освещения могут повлиять на восприятие цвета и деталей.

   • Искажения: Геометрические искажения, вызванные оптическими системами или движением объекта.

   • Разрешение: Ограниченное разрешение может привести к потере деталей.

   • Затенение и отражения: Могут скрывать детали объектов или создавать ложные сигналы.

   • Разнообразие объектов: Разные формы, размеры и цвета объектов могут затруднять их распознавание.


---



4. Влияние пространственных преобразований на изображение:

   • Пространственные преобразования (например, поворот, масштабирование, сдвиг) изменяют положение и форму объектов в изображении.

   • Они могут влиять на восприятие изображения, изменяя его геометрию и ориентацию.

   • Некоторые преобразования могут привести к потере информации или искажению деталей, особенно при интерполяции.

   • Корректное применение пространственных преобразований важно для задач, таких как выравнивание изображений и создание панорам.
